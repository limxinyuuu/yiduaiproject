```python
all_words=pd.Series(' '.join(tweet).lower().split())
word_count=pd.DataFrame(all_words.value_counts())
word_count.columns=['count']
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.figure(figsize=(16,8))
word_count.iloc[:20,:].plot(kind='bar')
plt.ylabel('Count Frequency')
plt.title('Most Frequent Words')
```

```python
wordcloud = WordCloud(width = 800, height = 800, 
                background_color ='white',  
                min_font_size = 10).generate(str(word_count)) 
plt.figure(figsize = (8, 8), facecolor = None) 
plt.imshow(wordcloud) 
plt.axis("off") 
plt.tight_layout(pad = 0) 
```

```python
#!pip install sklearn
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(tweet, df['label'], test_size=0.3,random_state=123)
```

```python
sentences=[ele.split(" ") for ele in X_train.values]
```

```python
from gensim.models import Word2Vec
model = Word2Vec(sentences, workers=4, size=50, sg=1, hs=0, window=2, min_count=2)
```

```python
vocabs=[]
vecs=[]
for word in model.wv.vocab:
    vocabs.append(word)
    vecs.append(model[word])
```

```python
from sklearn.manifold import TSNE
```

```python
sub_vecs=vecs[:1000]
sub_vocab=vocabs[:1000]
tsne_model=TSNE(perplexity=40, n_components=2, init='pca', n_iter=2000, random_state=23)
new_values=tsne_model.fit_transform(sub_vecs)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x=[]
y=[]
for value in new_values:
    x.append(value[0])
    y.append(value[1])
    
plt.figure(figsize=(12,8))
for i in range(len(x)):
    plt.scatter(x[i],y[i])
    plt.annotate(sub_vocab[i],xy=(x[i],y[i]),xytext=(2,2),textcoords='offset points', ha='right', va='bottom')
```
